```python
import numpy as np
import random
import matplotlib.pyplot as plt

import NeuralNet as skNN



def randGauss(nRand,mu,sigma):
    randArr=np.zeros((2,nRand)) # dim[0] = no. var, dim[1]=no. samp.
    for i in range(nRand):
        randArr[0,i]=random.gauss(mu=mu,sigma=sigma)
        randArr[1,i]=random.gauss(mu=mu,sigma=sigma)
    return randArr
```

## Generating random dataset in each class

```python
# generating data class 1
nSamp=50
p_1=randGauss(nSamp,3,2)
target_1=np.zeros((nSamp,1))

# generating data class 2
nSamp=50
p_2=randGauss(nSamp,11,2)
target_2=np.ones((nSamp,1))

# concat all data
p=np.concatenate((p_1,p_2),axis=1)
p=np.mat(p)
indexSampleP=1

target=np.concatenate((target_1,target_2),axis=0)
target=np.mat(target)
indexSampleTarget=0

# scatter plot data in each class
idxClass_1=np.array(target==0)
idxClass_2=np.array(target==1)
plt.plot(p[0,idxClass_1[:,0]],p[1,idxClass_1[:,0]],'or')
plt.plot(p[0,idxClass_2[:,0]],p[1,idxClass_2[:,0]],'ob')
plt.legend('Class_1','Class_2')
plt.show()
```

## Bulid the network

```python
# network structure
nHidLayer=1
nNodeInEachHid=np.array([1])
if len(nNodeInEachHid) != nHidLayer:
    print('You must set the no. of neural in each hidden layer !')

# array of transfer function that related with no. of hidden layer
# dim=len(nNeuronArr)-2 (set transfer funciton only neuron in each hidden layer)
transfNetwork=['hardlim']

# Build the network
nnSP1=skNN.MLP()
nnSP1.build(p,indexSampleP,target,indexSampleTarget,nNodeInEachHid,transfNetwork)

nnSP1.setWeightAndBias(1,30)
```

## Training

### Visualize step by text

```python
percError=10
nnSP1.train.unified(loopInInput=10,tol=percError,visualize='text',visualizeStep='none')
```
